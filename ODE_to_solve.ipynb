## The article
They used the fourth order Runge–Kutta finite difference method. They solved a known nonlinear system of differential equations:
$$
\frac{dy_1}{dt} = \cos{\left(t\right)} + y_1^{2} + y_2 - \left(1+t^{2}+\sin{\left(t\right)}^{2}\right), \\
\frac{dy_2}{dt} = 2t - \left(1+t^{2}\right)\sin{\left(t\right)} + y_1y_2,
$$
with $t \in [0,1]$ and $y_1\left(0\right)=0$, $y_2\left(0\right) = 1$. The analytic solution is:

$$
y_1 = \sin{\left(t\right)} \\
y_2 = 1+t^{2}
$$


## My examples

### Newton's Law of Cooling
$$
\frac{dT}{dt} = -k \left(T - A\left(t\right)\right)
$$
Analytic solution:
$$
T\left(t\right) = A + \left(T_0 -A\right)e^{-kt}.
$$

### Mixing problem
$$
\frac{dQ}{dt} + \frac{r\left(t\right)}{V}Q = q\left(t\right) r\left(t\right)
$$
Analytic solution (if $q, r$ are constants):
$$
Q\left(t\right) = qV-\left(qV-Q_0\right)e^{-\frac{q}{V}t}
$$

### Black–Scholes equation - but this is PDE, not ODE
$$
\frac{\partial V}{\partial t} + \frac{1}{2} \sigma^{2} S^{2} \frac{\partial^{2} V}{\partial S^{2}} + r S \frac{\partial V}{\partial S} - rV = 0
$$

Analytic solution:
$$
V\left(S,t\right) = SN\left(d_1\right) - K e^{-r \left(T-t\right)} N\left(d_2\right),
$$
where
$d_1 = \frac{\ln{\left(\frac{S}{K}\right)}+\left(r+\frac{\sigma^{2}}{2}\right)\left(T-t\right)}{\sigma \sqrt{T-t}}$, $d_2 = \frac{\ln{\left(\frac{S}{K}\right)}+\left(r-\frac{\sigma^{2}}{2}\right)\left(T-t\right)}{\sigma \sqrt{T-t}}$.

## System of ODEs
### 2-Body problem (or 3-Body problem - but here only numerical solution)
$$
m_i \frac{d^{2} x_i}{dt^{2}} = G \sum_{j=1, j \neq i} \frac{m_i m_j}{||x_j - x_i||^{2}} \frac{x_j - x_i}{||x_j - x_i||}
$$

Analytic solution for 2-Body problem:
$$
x_1\left(t\right) = R\left(t\right) + \frac{m_2}{m_1 + m_2} r\left(t\right) \\
x_2\left(t\right) = R\left(t\right) + \frac{m_1}{m_1 + m_2} r\left(t\right)
$$

### Lotka-Volterra predator-prey model (general version for n species) - but here only numerical solution
$$
\frac{dx}{dt} = \alpha x - \beta xy \\
\frac{dy}{dt} = -\gamma y + \delta xy
$$

### Different versions of SIR model
$$
\frac{dS}{dt} = -\frac{\beta IS}{N} \\
\frac{dI}{dt} = \frac{\beta IS}{N} - \gamma I \\
\frac{dr}{dt} = \gamma I
$$

The analytic solution:
$$
S\left(u\right) = S\left(0\right) u \\
I\left(u\right) = N - R\left(u\right) - S\left(u\right) \\
R\left(u\right) = R\left(0\right) - \rho \ln{\left(u\right)}
$$

## Final chosen ODEs to solve

### Newton's Law of Cooling - single ODE example
$$
\frac{dT}{dt} = -k \left(T - A\left(t\right)\right)
$$
Analytic solution:
$$
T\left(t\right) = A + \left(T_0 -A\right)e^{-kt}.
$$

### The SIS model - system of two ODEs example
$$
\frac{dS}{dt} = -\frac{\beta SI}{N} + \gamma I\\
\frac{dI}{dt} = \frac{\beta SI}{N} - \gamma I \\
$$
Analytics solution:
$$
I\left(t\right) = \frac{I_{\infty}}{1+V e^{-x t}} \\
S\left(t\right) = N - I\left(t\right),
$$
where $I_{\infty} = \left(1-\gamma/\beta\right)N$, $x = \beta - \gamma$, $V=I_{\infty}/I_{0}-1$

### Lorenz system - system of three ODEs exmample
$$
\frac{dx}{dt} = \sigma \left(y-x\right) \\
\frac{dy}{dt} = x\left(\rho - z\right) - y \\
\frac{dz}{dt} = xy - \beta x
$$

No analytic solution.
